```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from pandas.api.types import CategoricalDtype
```

```python
sns.set(rc={'figure.figsize':(8,6)})
```

```python
df_bnb=pd.read_csv("AB_NYC_2019.csv")
df_bnb.head(10)
```

```python
df_bnb["room_type"].value_counts().plot.bar().set_title("Room type frequencies");
#oda tiplerine göre sayılara veren barplot tablo oluşturduk
```

```python
df_bnb["room_type"].value_counts().plot.barh().set_title("Room type frequencies");
#yukarıdaki tablonun yatay hali
```

```python
sns.countplot(x="room_type", data=df_bnb, color="purple").set_title("room_type");
plt.show();
#daha komplike işlemler için seaborn kütüphanesini kullanabiliriz
#şu aşamada renk ve başlık özelliklerini kullandık
```

```python
sns.barplot(x="room_type", y=df_bnb["room_type"].index, data=df_bnb, color="purple").set_title("room_type")
```

```python
sns.barplot(x="minimum_nights",y="price", hue="room_type", data=df_bnb)
#burada anlamsız çıktı çünkü x ekseniçok kalabalık olduğu için mantıklı bir tablo oluşturulmadı
#bu anlamsızlığı çözmek için ilgi alanımız olan minimum gece verisini sınırlayabilriz
```

```python
#sınırlamak için pandas kütüphanesiyle veri setimizi güncelleyeceğiz
# Verilerinizi filtreleyin (3 ile 7 gün arasındaki minimum_nights değerleri)
filtered_df = df_bnb[(df_bnb["minimum_nights"] >= 3) & (df_bnb["minimum_nights"] <= 7)]
sns.barplot(x="minimum_nights", y="price", hue="room_type", data=filtered_df)
plt.title("Price by Minimum Nights and Room Type")
#böylece amacımıza ulaşabilriz belli syıda minimum gece içeren konaklama yerleri arasında oda tipine göre ücret karşılaştırması yapabiliriz
```

```python
#şehrlere göre pie charta yansıtalım
df_bnb.groupby("neighbourhood_group").sum().plot.pie(y="price",autopct="%.1f%%",ylabel="",legend=True,figsize=(10,10))
#sum price toplamlarının payını bulmamız için üst satırlarda varsayılan olarak ortalama kullandık
#burada figsizedam dolayı biraz sıkıştı bu yüzden biraz genişletelim
```

```python
cols=["neighbourhood_group","room_type"]
for col in cols:
  plt.figure(figsize=(10,10))
  sns.scatterplot(x=col,y="price",data=df_bnb)
  #price bağlı birden fazla grafik çizerken for döngüşü kullanabiliriz.
  #elmas da boyuta bağlı değerlendirme yapılmıştı burada ben neighborhood-price, roomtype-price a göre yapacağım
  #bunun çin catplot kullanmak daha doğru olur
```

```python
#buradan bunun için scatter plotun yanlış bir seçim olduğunu görüyoruz çünkü değerlerden biri kategorik
#bunun için birde number of views e göre bir lsite oluşturacağım
```

```python
plt.figure(figsize=(10,10))
sns.scatterplot(x="number_of_reviews",y="price",data=df_bnb)
#bunun için scatter plot daha iyi bir seçim oldu benzer şekilde avaliable days veya minimum nights gibi oranlar içinde scatter plot kullanabiliriz
```

```python
sns.scatterplot(x="minimum_nights",y="number_of_reviews",data=df_bnb)
#burada en sağda 1000 gecenin üstündeki veriler benim tablomu yanılttığı için onları eleyip tekrar oluşturacağım
```

```python
filtered_df = df_bnb[(df_bnb["minimum_nights"] <= 600)]
plt.figure(figsize=(10,10))
sns.scatterplot(x="minimum_nights",y="number_of_reviews",data=filtered_df)
#burada ihtyiaca göre daha küçültülebilir veya büyültülür veya farklı parametreler seçilebilir
```

```python
#az önce scatter plot ile gösterdiğimiz verileri histogram ile gösterleim
sns.histplot(x="minimum_nights",y="number_of_reviews", kde=True,data=filtered_df)
#terkar anlamsız çıktığı için veri setini biraz daha küçülteceğim
```

```python
filtered_df = df_bnb[(df_bnb["minimum_nights"] <= 100)]
sns.histplot(x="price",y="number_of_reviews", kde=True ,data=filtered_df, bins=150)
#bin genişliğini genişletip tekrar bakacağım
#kde gözle görülemiyor neden ? sanırım veri yapısı uygun olmadığı için kde çizgisi görüntülenemiyor.
```

```python
sns.jointplot(x="price", y=df_bnb["number_of_reviews"], data=df_bnb)
```

```python
#son grafik tiplerimize gelirken
sns.catplot(x="neighbourhood_group",y="price",kind="point", hue="room_type",data=df_bnb)
#catplotun bir başka kullanımı
```

```python
plt.figure.figsize=(10,10)
sns.FacetGrid(df_bnb, hue="room_type", height=5,xlim=(0,10000)).map(sns.kdeplot, "price", fill=True )
#verimi rahatça incelemediğim için pricea filtre uygulayıp tekrar deneyeceğim
```

```python
sns.FacetGrid(df_bnb, hue="room_type", height=5,xlim=(0,1000)).map(sns.kdeplot, "price", fill=True )
plt.figure.figsize=(10,5)
```
